```python
import pandas as pd
import os
import kagglehub
import numpy as np
```

```python
dataset_path = '/Users/geramukaeva/.cache/kagglehub/datasets/rohitsahoo/sales-forecasting/versions/2/train.csv'

if os.path.exists(dataset_path):
    superstore_dataset_raw = pd.read_csv(dataset_path)
superstore_dataset_raw.shape
```

преобразование данных

```python
superstore_dataset_raw = superstore_dataset_raw[
    ['Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 
     'City', 'State', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Sales']
    ]
superstore_dataset_raw.head()
```

```python
superstore_dataset_raw.rename(columns=str.lower, inplace=True)
superstore_dataset_raw.rename(columns={
    'row id': 'row_id',
    'order id': 'order_id',
    'order date': 'order_date',
    'ship date': 'ship_date',
    'ship mode': 'ship_mode',
    'customer id': 'customer_id',
    'customer name': 'customer_name',
    'product id': 'product_id',
    'sub-category': 'sub_category',
    'product name': 'product_name'
}, 
inplace=True)
```

```python
superstore_dataset_raw.isnull().sum()
```

```python
superstore_dataset_raw[superstore_dataset_raw.duplicated()].sort_values('row_id')
```

```python
superstore_dataset_raw['row_id'] = superstore_dataset_raw['row_id'].astype(object)
superstore_dataset_raw['order_date'] = pd.to_datetime(superstore_dataset_raw['order_date'], format='%d/%m/%Y')
superstore_dataset_raw['ship_date'] = pd.to_datetime(superstore_dataset_raw['ship_date'], format='%d/%m/%Y')
superstore_dataset_raw.info()
```

создание витрины

```python
superstore_mart = superstore_dataset_raw[
    ['row_id', 'order_id', 'order_date', 'ship_date', 'ship_mode', 'customer_id', 'customer_name', 'segment', 'city', 'state', 'region', 'category', 'sub_category', 'product_id', 'product_name', 'sales']
    ]
superstore_mart
```

нормализация датафрейма

```python
superstore_dataset_raw['order_date_id'] = pd.factorize(superstore_dataset_raw['order_date'])[0] + 1
superstore_dataset_raw['ship_date_id'] = pd.factorize(superstore_dataset_raw['ship_date'])[0] + 1
superstore_dataset_raw['ship_mode_id'] = pd.factorize(superstore_dataset_raw['ship_mode'])[0] + 1
superstore_dataset_raw['segment_id'] = pd.factorize(superstore_dataset_raw['segment'])[0] + 1
superstore_dataset_raw['city_id'] = pd.factorize(superstore_dataset_raw['city'])[0] + 1
superstore_dataset_raw['state_id'] = pd.factorize(superstore_dataset_raw['state'])[0] + 1
superstore_dataset_raw['region_id'] = pd.factorize(superstore_dataset_raw['region'])[0] + 1
superstore_dataset_raw['category_id'] = pd.factorize(superstore_dataset_raw['category'])[0] + 1
superstore_dataset_raw['sub_category_id'] = pd.factorize(superstore_dataset_raw['sub_category'])[0] + 1
```
